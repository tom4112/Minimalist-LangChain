# **LangChain Essentials**

### **What is LangChain?**

**LangChain** is a powerful tool designed from the ground up to help you build advanced AI applications. 

It gives you a consistent experience no matter which model you use behind the scenes and lets you switch between different AI vendors effortlessly.


You can use it to build:
* **Prompt Templates:** for easy formatting and maximum re-usability
* **Chains:** to connect multiple steps together (e.g. send a variable $\rightarrow$ get the prompt template filled-in $\rightarrow$ submit to the LLM $\rightarrow$ format response).
* **Tools:** for simple functions calling
* **AI Agents:** out of the box ReAct agents to run a LLM in a loop and call tools

---
### **Getting Started**
To use LangChain, you need the main framework plus the specific package for the AI provider you want to use.

### **1. Chat Models**

**LangChain** works differently than the official **OpenAI** or **Mistral AI** SDKs: 

* **The Native Way:** One client $\rightarrow$ multiple services (Chat, Embeddings, etc) for one vendor.
* **The LangChain Way:** Separate clients for each service (e.g. one Chat client, one Embedding client, etc) but easier to switch between vendors.

<br>
LangChain automatically looks for your API keys in your computer's environment variables.

If you are using local models via Ollama, you don't need a key at all, just make sure the Ollama is running!

| If you use this provider... | Install this package | Save your API Key as this name |
| :--- | :--- | :--- |
| **OpenAI** (ChatGPT) | `langchain-openai` | `OPENAI_API_KEY` |
| **Anthropic** (Claude) | `langchain-anthropic` | `ANTHROPIC_API_KEY` |
| **Mistral AI** | `langchain-mistralai` | `MISTRAL_API_KEY` |
| **Ollama** (Free / Local) | `langchain-ollama` | *Not required! (Run locally)* |

You can find the [list of all providers](https://docs.langchain.com/oss/python/integrations/chat#all-chat-models) here.

<br>

> There is also a universal `init_chat_model` option which we won't detail here as it's impossible to cover all options in LangChain... It still requires the installation of the different model providers packages.



```python
from langchain_mistralai import ChatMistralAI
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

# My current working directory includes a .env file with my API keys
load_dotenv()

# Let's set-up 2 clients from different vendors
mistral_llm = ChatMistralAI(model_name="mistral-small-latest")
openai_llm = ChatOpenAI(model_name="gpt-4o")
```

```python
# You can pass a string or a list of formatted messages such as [{"role": "user", "content": "hello"}]
mistral_llm.invoke("Hello!")
```

```python
# As you can see, the interface is consistent across all providers! The invoke method is also consistent across all LangChain 'runnables'
openai_llm.invoke("Hello!")
```

```python
# Langchain supports streaming: displaying text as it is being generated
# Use end="" and flush=True to avoid line breaks and force the notebook to display each chunk as soon as it is available
for chunk in mistral_llm.stream("What is LangChain?"):
    print(chunk.content, end="", flush=True)
```

```python
# You can use local models with Ollama
# You simply need to have Ollama running on your computer and choose a previouslydownloaded model

from langchain_ollama import ChatOllama

local_llm = ChatOllama(model="gemma4:latest", temperature=0)
local_llm.invoke("Hello")
```
