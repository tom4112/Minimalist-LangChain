# **LangChain Essentials**

### **8. Virtual Filesystem**

This middleware is part of **Deep Agents**, an open source harness developped by LangChain.

We will need to use:
- A [**backend**](https://reference.langchain.com/python/deepagents/backends/filesystem/FilesystemBackend) to choose the folder used to represent the root of our virtual filesystem
- The [**FilesystemMiddleware**](https://reference.langchain.com/python/deepagents/middleware/filesystem/FilesystemMiddleware) to expose this functionality to the agent and give it access to the backend

```python
%pip install deepagents
```

```python
from deepagents.backends import FilesystemBackend
from deepagents.middleware import FilesystemMiddleware

backend = FilesystemBackend(
    root_dir="./workspace", # When virtual_mode=True: Acts as a virtual root for filesystem operations
    virtual_mode=True       # When True, all paths are treated as virtual paths anchored to root_dir. Path traversal (.., ~) is blocked and all resolved paths are verified to remain within root_dir.
)
```

```python
from dotenv import load_dotenv
from langchain_mistralai import ChatMistralAI
from langchain.agents import create_agent
from langgraph.checkpoint.memory import MemorySaver

load_dotenv()

llm = ChatMistralAI(model="mistral-small-latest")

agent = create_agent(
    model=llm,
    checkpointer=MemorySaver(),
    middleware=[FilesystemMiddleware(backend=backend)] # without backend => files will be saved in the state
)
```

```python
from langchain_core.messages import HumanMessage

config = {"configurable": {"thread_id": "conversation1"}}

state = agent.invoke(
    {"messages": [HumanMessage(content="Write a poem about LangChain and save it in your workspace")]},
    config=config
)

state
```

```python
state = agent.invoke(
    {"messages": [HumanMessage(content="List all files available in your workspace")]},
    config=config
)

state
```
